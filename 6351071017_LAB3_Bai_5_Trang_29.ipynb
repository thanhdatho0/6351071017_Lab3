```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/thanhdatho0/6351071017_Lab3/main/Dataset/breast+cancer+wisconsin+diagnostic/wdbc.data", header=None)
df.columns = ["id", "diagnosis"] + [f"f{i}" for i in range(1, 30+1)]

X = df.drop(columns=["id", "diagnosis"])
y = df["diagnosis"].map({"B": 0, "M": 1})

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA 2 components
pca = PCA(n_components=2)
pca_2d = pca.fit_transform(X_scaled)

# Plot PCA1 vs PCA2 with true labels
plt.figure(figsize=(8, 6))
plt.scatter(pca_2d[y==0, 0], pca_2d[y==0, 1], label="Benign (Lành tính)")
plt.scatter(pca_2d[y==1, 0], pca_2d[y==1, 1], label="Malignant (Ác tính)")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("Hình ảnh PCA: Lành tính so với Ác tính")
plt.legend()
plt.tight_layout()
print("Hồ Thành Đạt - 6351071017")
plt.show()
```
