```python
STUDENT='Hồ Thành Đạt - 6351071017'
```

# Câu a

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler

iris = load_iris()
X = iris.data
y = iris.target
```

# Câu b

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

# Câu c

```python
print(STUDENT)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

def create_model():
    model = Sequential()
    model.add(Dense(10, input_shape=(4,), activation='relu'))
    model.add(Dense(20, activation='relu'))
    model.add(Dense(3, activation='softmax'))
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model
```

# Câu d

```python
print(STUDENT)

import os
import warnings
import logging
import numpy as np
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
warnings.filterwarnings("ignore")
logging.getLogger("tensorflow").setLevel(logging.ERROR)

kfold = KFold(n_splits=10, shuffle=True, random_state=42)
fold_no = 1

for train, test in kfold.split(X_scaled, y):
    X_train, X_test = X_scaled[train], X_scaled[test]
    y_train, y_test = y[train], y[test]

    model = create_model()
    model.fit(X_train, y_train, epochs=100, batch_size=10, verbose=0)

    y_pred = np.argmax(model.predict(X_test, verbose=0), axis=1)

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average='macro')
    rec = recall_score(y_test, y_pred, average='macro')
    f1 = f1_score(y_test, y_pred, average='macro')

    print(f"Fold {fold_no}: Accuracy={acc:.4f}, Precision={prec:.4f}, Recall={rec:.4f}, F1-score={f1:.4f}")

    model.save(f"model_fold_{fold_no}.h5")
    fold_no += 1
```

# Câu e

```python
print(STUDENT)

acc_scores = []
prec_scores = []
rec_scores = []
f1_scores = []

fold_no = 1

for train, test in kfold.split(X_scaled, y):
    X_train, X_test = X_scaled[train], X_scaled[test]
    y_train, y_test = y[train], y[test]

    model = create_model()
    model.fit(X_train, y_train, epochs=100, batch_size=10, verbose=0)

    y_pred = np.argmax(model.predict(X_test, verbose=0), axis=1)

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average='macro')
    rec = recall_score(y_test, y_pred, average='macro')
    f1 = f1_score(y_test, y_pred, average='macro')

    acc_scores.append(acc)
    prec_scores.append(prec)
    rec_scores.append(rec)
    f1_scores.append(f1)

    print(f"Fold {fold_no}: Accuracy={acc:.4f}, Precision={prec:.4f}, Recall={rec:.4f}, F1-score={f1:.4f}")

    model.save(f"model_fold_{fold_no}.h5")
    fold_no += 1

print("-" * 50)
print(f"Average Accuracy: {np.mean(acc_scores):.4f}")
print(f"Average Precision: {np.mean(prec_scores):.4f}")
print(f"Average Recall: {np.mean(rec_scores):.4f}")
print(f"Average F1-score: {np.mean(f1_scores):.4f}")
```

# Câu f

```python
print(STUDENT)

cls_precisions = []
cls_recalls = []
cls_f1s = []

fold_no = 1

for train, test in kfold.split(X_scaled, y):
    X_train, X_test = X_scaled[train], X_scaled[test]
    y_train, y_test = y[train], y[test]

    model = create_model()
    model.fit(X_train, y_train, epochs=100, batch_size=10, verbose=0)

    y_pred = np.argmax(model.predict(X_test, verbose=0), axis=1)

    # average=None để lấy chỉ số cho từng class
    cls_precisions.append(precision_score(y_test, y_pred, average=None))
    cls_recalls.append(recall_score(y_test, y_pred, average=None))
    cls_f1s.append(f1_score(y_test, y_pred, average=None))

    print(f"Fold {fold_no} completed.")
    fold_no += 1

# Tính trung bình theo cột (axis=0)
avg_prec_per_class = np.mean(cls_precisions, axis=0)
avg_rec_per_class = np.mean(cls_recalls, axis=0)
avg_f1_per_class = np.mean(cls_f1s, axis=0)

print("-" * 60)
class_names = iris.target_names
for i, name in enumerate(class_names):
    print(f"Class '{name}':")
    print(f"  Average Precision: {avg_prec_per_class[i]:.4f}")
    print(f"  Average Recall:    {avg_rec_per_class[i]:.4f}")
    print(f"  Average F1-score:  {avg_f1_per_class[i]:.4f}")
    print("-" * 60)
```

```python
print(STUDENT)

from tensorflow.keras.models import load_model

# 1. Xác định mô hình tốt nhất
best_index = np.argmax(f1_scores) # Lấy index của fold có F1 cao nhất
best_fold = best_index + 1
print(f"Mô hình tốt nhất: Fold {best_fold} (F1-score: {f1_scores[best_index]:.4f})")

# Load mô hình
best_model = load_model(f"model_fold_{best_fold}.h5")

# 2. Chuẩn bị dữ liệu mới
new_data = np.array([
    [6.2, 2.9, 4.3, 1.3],
    [5.1, 3.5, 1.4, 0.2],
    [7.3, 2.8, 6.4, 2.1]
])

# Chuẩn hóa dữ liệu mới bằng scaler cũ (quan trọng)
new_data_scaled = scaler.transform(new_data)

# 3. Dự đoán
preds = np.argmax(best_model.predict(new_data_scaled, verbose=0), axis=1)
class_names = iris.target_names

print("-" * 50)
for i, pred in enumerate(preds):
    print(f"Dữ liệu {new_data[i]} => Phân lớp: {class_names[pred]}")
```
